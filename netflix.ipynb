<center>
<h1> Load the Data and the Exploration

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data=pd.read_csv("netflix_titles.csv")
data.head()
```

```python
data.info()
```

<<center>>
<h1> Clean The Data

```python
data=data.dropna(subset=['type','country','release_year','rating','duration'])
data.head(10)
```

```python
count=data['type'].value_counts()
count
```

<center>
<h1> TVs Vs Movies Show

```python
plt.Figure(figsize=(6,4))
plt.bar(count.index,count.values,color=["Red","Black"])
plt.title("Number Of Movies vs TV's Shows on Netflix")
plt.ylim(0,7000)
plt.xlabel("Type")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("Movies_Vs_TvShows.png")
plt.show()
```

<center>
<h1> Rating 



```python
rating_counts=data['rating'].value_counts()
rating_counts
```

```python
plt.Figure(figsize=(15,8))
plt.pie(rating_counts,labels=rating_counts.index,autopct='%1.1f%%',labeldistance=1.2)
plt.title("Count the Differnet Types of rating")
plt.savefig("Count_rating.png")
plt.show()
```

<center>
<h1> Duration

```python
movies=data[data['type']=='Movie'].copy()
movies['duration']=movies['duration'].str.replace("min","").astype(float)

plt.Figure(figsize=(8,6))
plt.hist(movies["duration"],bins=30,color="lightgreen",edgecolor="black")
plt.xlabel("Minutes")
plt.ylabel("No of Movies")
plt.title("Distribution Of Movie Duration ")
plt.tight_layout()
plt.savefig("Distribution_Of_Movie_Duration_")
plt.show()
```

```python
realease_counts=data['release_year'].value_counts().sort_index()
realease_counts
```

```python
plt.Figure(figsize=(18,15))
plt.scatter(realease_counts.index,realease_counts.values,color="red",)
plt.xlabel("Release Year")
plt.ylabel("No of shows")
plt.title("Release Year Vs No of Shows")
plt.tight_layout()
plt.savefig("realease_year_scatte.png")
plt.show()
```

<center>
<h1>country count

```python
country=data['country'].value_counts().head(10)
plt.Figure(figsize=(18,15))
plt.barh(country.index,country.values,color="orange",)
plt.xlabel("No of Shows")
plt.ylabel("Country")
plt.title("Top 10 countries")
plt.tight_layout()
plt.savefig("toptencountries.png")
plt.show()
```

<center>
<h1>Movies and TV shows per Year

```python



content = data.groupby(["release_year","type"]).size().unstack().fillna(0)

fig, ax = plt.subplots(1,2,figsize=(12,5))


ax[0].plot(content.index, content["Movie"], color="blue", marker="o")
ax[0].set_title("Movies Released Per Year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("No. of Movies")


ax[1].plot(content.index, content["TV Show"], color="red", marker="o")
ax[1].set_title("TV Shows Released Per Year")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("No. of TV Shows")


fig.suptitle("Movies and TV Shows Released by Year", fontsize=14)
plt.tight_layout()
plt.savefig('Movies_and_TVShows_per_Year.png')
plt.show()
```
